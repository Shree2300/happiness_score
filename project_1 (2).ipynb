```python
economy_happiness = ['Happiness_Score','Economy(GDP_per_Capita)']
economy_corr = df[economy_happiness]
economy_corr.corr()
```

```python
#correlation graph

sns.regplot(data = economy_corr, x = 'Economy(GDP_per_Capita)', y = 'Happiness_Score').set_title("Correlation graph for Happiness_score vs Economy")
```

```python
#Creating R and R Squared for Happiness Score and Economy

R = np.array(economy_corr.corr())
R2 = np.array(economy_corr.corr()**2)

print('The data shows we have an R value of: ' + str(R[1]) + ' and and R2 value of ' + str(R2[1]))
```
